```python
myStudents = ['juan', 'pedro', 'sofia', 'karen', 'pepita']
```

```python
for student in myStudents:
    if 'a' in student:
        print('''this student's name contains an "a"''')
    else:
        print('it does not contain an "a"')
    print('helo')
```

```python
for student in myStudents:
    if 'a' in student:
        print('''this student's name contains an "a"''')
    else:
        print('it does not contain an "a"')
print('helo')
```

```python
for student in myStudents:
    if 'a' in student:
        print(student + "'s name contains an 'a'")
    elif 'e' in student:
        print(student + "'s name does not contain an 'a' but contains an 'e'")
    else:
        print(student + "'s name does not contain an 'a' or and 'e'")
```
