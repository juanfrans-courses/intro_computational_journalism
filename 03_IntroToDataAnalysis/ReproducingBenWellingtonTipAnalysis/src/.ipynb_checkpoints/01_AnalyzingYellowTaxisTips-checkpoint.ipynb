# Reproducing Ben Wellington's analysis of NYC taxi tips

This notebook reproduces Ben Wellington's analysis of New York City taxi tips in 2013. It is based on his article [How software in half of NYC cabs generates $5.2 million a year in extra tips](https://iquantny.tumblr.com/post/107245431809/how-software-in-half-of-nyc-cabs-generates-52). It uses the full month of yellow [taxi data](https://www1.nyc.gov/site/tlc/about/tlc-trip-record-data.page) (October 2013), but it can also be replicated with smaller datasets (eg. one week).

**Creating a histogram of the tip percentage**

```python
taxiData['tip_perc'] = taxiData['tip_amount'] / (taxiData['fare_amount'] + taxiData['surcharge'])
```

```python
bins = np.arange(0, .41, 0.01)
plt.rcParams["figure.figsize"] = [10,5]
plt.hist(taxiData['tip_perc'], bins=bins)
plt.show()
```

**Selecting only the trips paid with credit card**

```python
creditPayment = taxiData[taxiData['payment_type'] == 'CRD']
```

```python
creditPayment.head()
```

```python
bins = np.arange(0, .41, 0.01)
plt.rcParams["figure.figsize"] = [10,5]
plt.hist(creditPayment['tip_perc'], bins=bins)
plt.show()
```

```python
creditPayment['tip_perc2'] = creditPayment['tip_amount'] / (creditPayment['fare_amount'] + creditPayment['surcharge'] + creditPayment['mta_tax'] + creditPayment['tolls_amount'])
```

```python
creditPayment.head()
```

```python
creditPayment['trip_perc3'] = np.where(creditPayment['vendor_id'] == 'CMT', creditPayment['tip_perc2'], creditPayment['tip_perc'])
```

```python
taxiData['vendor_id'].unique()
```

```python
bins = np.arange(0, .41, 0.01)
plt.rcParams["figure.figsize"] = [10,5]
plt.hist(creditPayment['trip_perc3'], bins=bins)
plt.show()
```

```python
creditPayment['trip_perc3'].value_counts(ascending=False)
```

There are two different 0.200000 values. We need to merge them together.

```python
creditPayment['roundedTip'] = creditPayment['trip_perc3'].round(4)
```

```python
creditPayment['roundedTip'].value_counts(ascending=False)
```

```python
bins = np.arange(0, .41, 0.01)
plt.rcParams["figure.figsize"] = [10,5]
plt.hist(creditPayment['roundedTip'], bins=bins)
plt.show()
```

**Calculating how much extra tips people have paid**

```python
creditPayment['shouldHavePaid'] = np.where(creditPayment['vendor_id'] == 'CMT', creditPayment['tip_perc2'] * (creditPayment['fare_amount'] + creditPayment['surcharge']), creditPayment['tip_amount'])
```

```python
creditPayment.head()
```

```python
creditPayment['extraPaid'] = creditPayment['tip_amount'] - creditPayment['shouldHavePaid']
```

```python
creditPayment['extraPaid'].sum()
```

**Conclusion**: in October 2013 users paid $621,188 extra dollars in tips due to how vendor 'CMT' calculates its default values. This is probably slightly lower because not all people used the default value. This calculation should probably be refined a bit more to look a the difference in payment of **only** those users that chose 15%, 20%, 25%, or 30%.
