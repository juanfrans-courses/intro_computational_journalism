# Reproducing Ben Wellington's analysis of NYC taxi tips

**Import libraries: Pandas, Numpy, Matplotlib**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Import one month of taxi data**

```python
taxiData = pd.read_csv('../data/yellow_tripdata_2014-10.csv', delimiter=',', header=0, encoding='UTF-8')
```

```python
taxiData.head()
```

```python
taxiData.shape
```

**Calculate new column for tip percentage**

```python
taxiData['tip_perc'] = taxiData[' tip_amount'] / (taxiData[' fare_amount'] + taxiData[' surcharge'])
```

```python
taxiData.head()
```

**Create histogram of the tip percentage**

```python
bins = np.arange(0, 0.41, 0.01)
plt.hist(taxiData['tip_perc'], bins=bins)
plt.show()
# Change size
# Remove the 0%
```
